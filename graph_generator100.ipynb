```python
import networkx as nx 
import matplotlib.pyplot as plt 
import numpy as np
```

```python
import random
for x in range(200):
    print(random.randint(1,30))
```

```python
G= nx.erdos_renyi_graph(200,0.03) 
nx.draw(G, with_labels=True)
plt.show() 
```

```python
matrix = nx.adjacency_matrix(G)
np_matrix = matrix.toarray()
print(np_matrix)
```

```python
np.savetxt('dummy_graph.txt',np_matrix,fmt='%i', delimiter=' ')
```
