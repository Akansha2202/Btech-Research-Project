```python
import networkx as nx 
import matplotlib.pyplot as plt 
```

```python
import numpy as np
```

## Erdosâ€“Renyi.txt

```python
G= nx.erdos_renyi_graph(20,0.2) 
nx.draw(G, with_labels=True) 
plt.show() 
```

```python
A = nx.adjacency_matrix(G)
```

```python
A.todense()
```

```python
matrix = nx.adjacency_matrix(G)
np_matrix = matrix.toarray()
print(np_matrix)
```

```python
np.savetxt('Erdosâ€“Renyi.txt',np_matrix,fmt='%i', delimiter=' ')
```

## Degree Centrality

```python
deg_centrality = nx.degree_centrality(G) 
#print(deg_centrality)
```

```python
dictDEG={k: v for k, v in sorted(deg_centrality.items(), key=lambda item: item[1])}
```

```python
print(dictDEG)
```

```python

try: 
	geeky_file = open('ErdosRenyi/deg_centrality.txt', 'wt') 
	geeky_file.write(str(dictDEG)) 
	geeky_file.close() 

except: 
	print("Unable to write to file")
```

## Closeness Centrality 

```python
close_centrality = nx.closeness_centrality(G) 
print(close_centrality)
```

```python
dictCLO={k: v for k, v in sorted(close_centrality.items(), key=lambda item: item[1])}
```

```python
try: 
	geeky_file = open('ErdosRenyi/clo_centrality.txt', 'wt') 
	geeky_file.write(str(dictCLO)) 
	geeky_file.close() 

except: 
	print("Unable to write to file")
```

## Betweenness Centrality

```python
bet_centrality = nx.betweenness_centrality(G, normalized = True,  
                                              endpoints = False) 
#print(bet_centrality)
```

```python
dictBET={k: v for k, v in sorted(bet_centrality.items(), key=lambda item: item[1])}
```

```python
try: 
	geeky_file = open('ErdosRenyi/bet_centrality.txt', 'wt') 
	geeky_file.write(str(dictBET)) 
	geeky_file.close() 
except: 
	print("Unable to write to file")
```

## Page Rank

```python
page_rank = nx.pagerank(G, alpha = 0.8) 
print(page_rank)
```

```python
dictPG={k: v for k, v in sorted(page_rank.items(), key=lambda item: item[1])}
```

```python
try: 
	geeky_file = open('ErdosRenyi/page_rank.txt', 'wt') 
	geeky_file.write(str(dictPG)) 
	geeky_file.close() 
except: 
	print("Unable to write to file")
```
